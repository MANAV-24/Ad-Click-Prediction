```python
import pandas as pd
B=pd.read_csv("ad_click_dataset.csv")
print(B.to_string)
```

```python
B.columns
```

```python
B.dropna(inplace=True)
```

```python
B["device_type"].unique()
```

```python
print(B.to_string)
```

```python
B["ad_position"].unique()
```

```python
B["browsing_history"].unique()
```

```python
B["time_of_day"].unique()
```

```python
B["ad_position"].replace({"Bottom":0,"Top":1,"Side":2},inplace=True)
B["device_type"].replace({"Tablet":0,"Mobile":1,"Desktop":2},inplace=True)
B["browsing_history"].replace({"News":0,"Education":1,"Social Media":2,"Shopping":3,"Entertainment":4},inplace=True)
B["time_of_day"].replace({"Morning":0,"Afternoon":1,"Evening":2,"Night":3},inplace=True)
B["gender"].replace({"Male":0,"Female":1,"Non-Binary":2},inplace=True)
```

```python
print(B.to_string())
```

```python
X=B[["age","gender","device_type","ad_position","time_of_day","browsing_history"]]
Y=B["click"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=150,random_state=1)
RF.fit(X_train,Y_train)
Y_pred=RF.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(accuracy_score(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python
import pickle

# Save the trained model
with open("ad_click_model.pkl", "wb") as model_file:
    pickle.dump(RF, model_file)

print("Model saved successfully!")
```
